```python
df=pd.read_csv("/kaggle/input/summeranalytics2020/train.csv",index_col="Id")
df2=pd.read_csv("/kaggle/input/summeranalytics2020/test.csv",index_col="Id")
df.dtypes
```

```python
categorical_feature_mask=df.dtypes==object
categorical_cols=df.columns[categorical_feature_mask].tolist()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
for col in categorical_cols:
    le.fit(df[col])
    df[col]=le.transform(df[col])
    df2[col]=le.transform(df2[col])
```

```python
y_train=df["Attrition"]
df=df.drop(columns=["Attrition","EmployeeNumber","Behaviour"])
df2=df2.drop(columns=["EmployeeNumber","Behaviour"])
```
